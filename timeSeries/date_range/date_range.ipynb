```python
rng = pd.date_range(start="6/1/2017", end="6/30/2017", freq='B') #exclude saturday and sunday.consider Business days
rng
```

```python
df.asfreq('D', method='pad') # create table with weekends, 
#pad method just carry forward data to next day which was missing
```
