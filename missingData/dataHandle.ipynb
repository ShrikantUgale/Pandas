```python
def replace_temp(cell):
    if not cell:
        return 0
    return cell

df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'], converters = {
    'temperature':replace_temp
})

df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df.fillna(0) #replace all NaN values with 0
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})   #replace NaN values specific to column
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.fillna(method='ffill')  # forward fill (carry forward current day valu to next)
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.fillna(method='bfill')  # backward fill vertially
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.fillna(method='bfill', axis='columns')  # backward fill (next columns value copied in previous column)
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.fillna(method='ffill', limit=1)  #ffill till limit 1, that means if next 2 values are blank then fill only 1 value
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.interpolate()
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.interpolate(method = 'time') #more near according to date
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.dropna() # drop all rows with at least 1 NaN
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.dropna(how='all') # drop rows where all columns contain NaN
df
```

```python
df = pd.read_excel('weather_data.xls', 'weather_data', parse_dates=['day'])
df.set_index('day', inplace=True)
df = df.dropna(thresh=2) # if I have at least 2 non NaN value in row then keep that row
df
```
